```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/france_energy_consumption.csv')
# BP Statistical Review of World Energy
#df['Oil Consumption - EJ']
df['Biofuels (TWh)'] = df['Biofuels (TWh)'] / 277.778
df = df.rename(columns={'Biofuels (TWh)': 'Biofuels - EJ'})
df['total'] = df.iloc[:, 3:].sum(axis=1)
df

df['Oil Consumption - EJ'] /= df['total']
df['Gas Consumption - EJ'] /= df['total']
df['Coal Consumption - EJ'] /= df['total']
df['Solar Consumption - EJ'] /= df['total']
df['Hydro Consumption - EJ'] /= df['total']
df['Nuclear Consumption - EJ'] /= df['total']
df['Wind Consumption - EJ'] /= df['total']
df['Geo Biomass Other - EJ'] /= df['total']
df['Biofuels - EJ'] /= df['total']
df = df.rename(columns={'Oil Consumption - EJ': 'Oil Consumption - %',
                        'Gas Consumption - EJ': 'Gas Consumption - %',
                        'Coal Consumption - EJ': 'Coal Consumption - %',
                        'Solar Consumption - EJ': 'Solar Consumption - %',
                        'Hydro Consumption - EJ': 'Hydro Consumption - %',
                        'Nuclear Consumption - EJ': 'Nuclear Consumption - %',
                        'Wind Consumption - EJ': 'Wind Consumption - %',
                        'Geo Biomass Other - EJ': 'Geo Biomass Other - %',
                        'Biofuels - EJ': 'Biofuels Consumption - %',
                        
                        })
df['total - %'] = df.iloc[:, 3:-1].sum(axis=1)
df
```

```python
# Installed Capacity

# import matplotlib.pyplot as plt

# def my_autopct(pct):
#     return ('%1.1f%%' % pct) if pct > 3 else ''


# # def my_label(labels):
# #     return () if pct > 3 else ''
# # Pie chart
# labels = ['Nuclear', 'Hydroelectric', 'Wind', 'Natural Gas', 'Solar', '', 'Other', '']
# sizes = [45.1, 18.9, 13.0, 9.3, 7.6, 2.5, 2.1, 1.6]

# #colors
# colors = ['#99ff99','#66b3ff','#95d7e6', '#ff9999','#ffcc99', '#944a2f', '#717275','#b8ac88']
 
# fig1, ax1 = plt.subplots()
# # explode = (0.05,0.05,0.05,0.05,0.05,0.05, 0.05, 0.05)
# explode = (0,0,0,0,0,0,0,0)
# ax1.pie(sizes, colors=colors, labels=labels, autopct=my_autopct, pctdistance=0.85, startangle=90, explode=explode)
# #draw circle
# # plt.legend(labels, loc='center left', bbox_to_anchor=(-0.1, 1.),
# #            fontsize=8)
# centre_circle = plt.Circle((0,0),0.70,fc='white')
# fig = plt.gcf()
# fig.gca().add_artist(centre_circle)
# # Equal aspect ratio ensures that pie is drawn as a circle
# ax1.axis('equal')  
# plt.tight_layout()
# plt.savefig('installed_capacity.png')

#S2b

fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))

production = ['Oil', 'Natural Gas', 'Coal', 'Solar', 'Hydro', 'Nuclear', 'Wind', 'Biomass']
data = [0.329606, 0.163647, 0.028225, 0.010943, 0.054595, 0.372655, 3.216835e-02, 0.008114]
radius = 1.



wedges, texts = ax.pie(data, wedgeprops=dict(width=0.2), radius=radius, startangle=-40, colors=['#b8ac88', '#ff9999', '#944a2f', '#ffcc99', '#66b3ff', '#99ff99', '#95d7e6', '#717275', '#717275'])

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(production[i], xy=(x, y), xytext=(1.0*np.sign(x), 1.1*y),
                horizontalalignment=horizontalalignment, **kw)


plt.tight_layout()
plt.savefig('france_energy_mix_2019.png')
```
