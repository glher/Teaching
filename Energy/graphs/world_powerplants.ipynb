```python
import pandas as pd
```

```python
df = pd.read_csv('data/globalpowerplantdatabasev120/global_power_plant_database.csv')
# print(list(df['primary_fuel'].unique()))
sources = ['Hydro', 'Gas', 'Oil', 'Wind', 'Nuclear', 'Coal', 'Solar', 'Waste', 'Biomass', 'Wave and Tidal', 'Geothermal']
# ['Petcoke', 'Geothermal', 'Cogeneration', 'Storage', 'Other']
for source in sources:
    installed_capacity = df[df['primary_fuel'] == source]['capacity_mw'].sum()
    generation = df[df['primary_fuel'] == source]['estimated_generation_gwh'].sum()
    
    print(f'{source}: {1e-3*installed_capacity:.1f} TWe, {1e-3*generation:.1f} TWh')
# df[df['primary_fuel'] == 'Storage']['capacity_mw'].sum()
```

```python
# https://www.bp.com/content/dam/bp/business-sites/en/global/corporate/pdfs/energy-economics/statistical-review/bp-stats-review-2020-full-report.pdf
oil: 53619.5  --> Primary energy
gas: 39291.7  --> Primary energy
coal: 43850  --> Primary energy
# https://ourworldindata.org/nuclear-energy
nuclear: 2616.6  --> Electricity
# https://ourworldindata.org/renewable-energy
hydro: 4355  --> Electricity
solar: 844.4  --> Electricity
wind: 1590.2  --> Electricity

We want to transition oil, gas and coal. Doing so will require less energy to install
BP gives a efficiency factor of around 40%. This means that 40% of the primary energy is converted to electricity. TO generate X TWh, X = 40% * Y, Y is the value given by BP in oil-equivalent. So, to go from fossil fuel Y consumption to electricity consumption efficiency

Nuclear heat rate: 10500 (33%)
Oil: 20-30%
EVs: 75%

So for 53619.5 TWh of oil (assuming in 25% efficient machines), this translates to 13405 TWh useful
For a 75% efficiency of EVs, 13405 / 0.77 = 17409 TWh

The fossil fuel equivalency adjustment represents the energy that would have been consumed if electricity had been
generated by fossil fuels.
Assuming an efficiency factor of around 40%, this means that Electricity / 40% = Primary energy equivalent
So, say for hydro, E = 4355 TWh. This means that to back it up, one would need 4355 / 0.4 = 10888 TWh of fossil fuel
Alternatively, to replace 53619.5 TWh of fossil fuel, one would need 53619.5 * 0.4 = 21447 TWh
If a gasoline powered car is 30% efficient, and an EVs 75% efficient, then the equivalency factor is 0.30 / 0.75 = 0.4 for cars

Transitions:

oil to renewable: 53619.5 * 0.4 = 21447 TWh
gas to renewable: 39291.7 * 0.4 = 15717 TWh
coal to renewable: 43850 * 0.4 = 17540 TWh
nuclear: 2616.6 TWh
hydro: 4355 TWh
solar: 844.4 TWh
wind: 1590.2 TWh

Total transition:
146167 TWh ----> 64111 TWh

At 200000 TWh ----> 87690 TWh

Kind of optimistic
```
