```python
#S1a

fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))

production = ["100% Wind - Onshore"]
data = [100]
radius = 1.

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.2), radius=radius, startangle=-40, colors=['#95d7e6'])

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(production[i], xy=(x, y), xytext=(1.05*np.sign(x), 1.2*y),
                horizontalalignment=horizontalalignment, **kw)



storage = ["100% Pumped Hydro"]
data = [100]
radius = 0.8

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.2), radius=radius, startangle=40, colors=['#66b3ff'])

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=1, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang)) * radius
    x = np.cos(np.deg2rad(ang)) * radius
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(storage[i], xy=(x, y), xytext=(1.15*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

plt.tight_layout()
plt.savefig('S1a.png')
```
